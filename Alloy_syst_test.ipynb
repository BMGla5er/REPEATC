```python
user_database = False
#user_path = 'Jupyter/Al_jan05.TDB'
tc_database = "TCNI12"

dependent_element = "Fe"
composition = {"Ni": 0.535,
               "Cr": 0.1946,
               "Nb": 0.0531,
               "Al": 0.0031,
               "Ti": 0.01,
               "Co": 0.002,
               "Mo": 0.0299
               #"C": 0.00034
               #"O":  0.000574441
               }  # in weight fraction #
comp_metric = 'W' # W: weight fraction, X: mole fraction

param = 'T'
param_range = range(300,1300,100) #in Kelvin
Temp = 923.15 #kelvin  = 650 degree C

elements = list(composition.keys())
nelem = len(elements)
# suspending_phases = ['BCC_B2#1','SIGMA#1','NI3TA_D0A#1', 'C14_LAVES#1','M6C#1','R_PHASE#1']
suspending_phases = ['NI3TA_D0A#1', 'NI3TI_D024#1','P_PHASE#1']
excluded_keys = ["BCC_B2#1","BCC_B2#3", "FCC_L12#4"]
```

```python
alloy = alloy_sys(dependent_element,composition,param, param_range ,user_database,tc_database,comp_metric,suspending_phases)

alloy.do_perform_single_axis_split(T=923.15,threshold=0.0001)
#do_peform_single_axis_split(alloy,T=923.15,threshold=0.001)
```

```python
alloy.phase_distribution()
alloy.composition_distribution()
alloy.element_distribution()
```
